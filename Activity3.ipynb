```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data.head()
```

```python
df=pd.DataFrame(data)
```

```python
df
```

# 1) Compare churn count with respect to gender.

```python
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df, ['gender', 'Churn'],title='Churn count with respect to gender')
```

```python
sns.countplot(data=df,x='gender',hue='Churn')
plt.title('Churn count with respect to gender')
plt.xlabel('Gender')
plt.show()
```

# 2) Findout how many female senior citizen there in the dataset

```python
df1.groupby(['gender']).sum()
```

# 3) Compare 'tenure' with 'Total Charges'

```python
df2=pd.DataFrame(data,columns=['tenure','TotalCharges'])
```

```python
df2
```

```python
df2['TotalCharges']=pd.to_numeric(df2['TotalCharges'],errors='coerce')
df2['tenure']=pd.to_numeric(df2['tenure'],errors='coerce')
x=df2['tenure']
y=df2['TotalCharges']
plt.figure(figsize=(10,10))
plt.bar(x,y)
plt.title('Tenure v/s Total charges',fontsize=15)
plt.xlabel('Tenure',fontsize=12)
plt.ylabel('Total Charges',fontsize=12)
plt.show()
```
